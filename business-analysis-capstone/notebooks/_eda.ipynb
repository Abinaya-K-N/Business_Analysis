Loading CLEANED data

```python
import pandas as pd
import matplotlib.pyplot as plt

sales = pd.read_csv("../data/cleaned_sales.csv")
house = pd.read_csv("../data/cleaned_house.csv")
churn = pd.read_csv("../data/cleaned_churn.csv")
```

Visualizations

Sales Distribution

```python
sales["Total_Sales"].hist()
plt.title("Sales Distribution")
plt.show()
```

Sales by Region

```python
sales.groupby("Region")["Total_Sales"].sum().plot(kind="bar")
plt.show()
```

House Price Distribution

```python
house["Price"].hist()
plt.show()
```

Area vs Price

```python
plt.scatter(house["Area"], house["Price"])
plt.show()
```

Churn Count

```python
churn["Churn"].value_counts().plot(kind="bar")
plt.show()
```
